```python
# 6) Quel est l’item le plus commandé 
# plt.figure(figsize=(20,6))

# plt.bar(resto_data.item_name, resto_data.quantity)

resto_data_grouped = resto_data.groupby("item_name").sum().sort_values("quantity", axis = 0, ascending=False)

resto_data_grouped
```

```python
# 7) Combien d’items ont été commandés au total chez Chipotle
print(sum(resto_data_grouped.quantity))

print(sum(resto_data.quantity))
```

```python
# 8) Combien de chiffre d’affaire à realisé Chipotle ? 
# sum(resto_data.item_price)
CA_Chipotle = 0.0

# TODO : iter over copy of colomns
for index, row in resto_data.iterrows():
    CA_Chipotle += (float(row['item_price'][1:])*float(row['quantity']))

CA_Chipotle
```

```python
# 9) Quel est le revenu moyen par commande ?
moyCommande = CA_Chipotle/len(resto_data.order_id.unique())

moyCommande
```
